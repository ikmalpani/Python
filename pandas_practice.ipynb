```python
#filter rows of a pandas DataFrame by column value
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.shape
```

```python
#long way
#using boolean to create a python list with same length as dataframe
#true only if the filter condition is satisfied- false otherwise
booleans = []
for length in movies.duration:#pandas series are iterable- like lists
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
#convert booleans list to pandas series
is_long = pd.Series(booleans)
```

```python
is_long.head() #series have a head method just like dataframes
```

```python
movies[is_long] #only true for the filter condition are shown 
```

```python
#short way
is_long = movies.duration >= 200 #comparision orperator - series comparision
#element wise comparision - return a series of true or false
is_long.head()
```

```python
movies[is_long] #pass the series
```

```python
#Do all together - used more often - for comparision
movies[movies.duration >= 200]
```

```python
#if only one column req
movies[movies.duration >= 200].genre #or use bracket notation
```

```python
# using loc method
movies.loc[movies.duration >= 200,'genre'] #dataframe.loc[rows,columns]
```

```python
#SettingWithCopyWarning in pandas
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.content_rating.isnull().sum()
```

```python
movies[movies.content_rating.isnull()]
```

```python
movies.content_rating.value_counts()#need to replace not rated as NaN
```

```python
movies[movies.content_rating=='NOT RATED']
```

```python
movies[movies.content_rating=='NOT RATED'].content_rating
```

```python
movies[movies.content_rating=='NOT RATED'].content_rating = np.nan
#warning because
#movies[movies.content_rating=='NOT RATED'] - first operation -get item
#.content_rating = np.nan - second operation - set item
#get item can return a view or a copy in pandas
#view is okay but copy is not but pandas doesnt know if it is a view or a copy
```

```python
movies.content_rating.isnull().sum()#didnt work
```

```python
movies.loc[movies.content_rating=='NOT RATED','content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()#works
```

```python
#example 2
top_movies = movies.loc[movies.star_rating >= 9,:]
top_movies
```

```python
top_movies.loc[0,'duration'] = 150
#same warning even after using loc
#because pandas doesnt know if it is a view or a copy to the original movies df
```

```python
top_movies#warning but still worked
```

```python
#solution - use .copy() explicitly
top_movies = movies.loc[movies.star_rating >= 9,:].copy()
top_movies
```

```python
top_movies.loc[0,'duration'] = 150
top_movies
```
